SERIES IN PANDAS

```python
data=[1,2,3,4,5]
series=pd.Series(data)
print(series)
print(type(series))
```

```python
#creating an series from  dcitionary
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
strr=[10,40,50]
ind=['a','b','c']
pd.Series(strr,index=ind)
```

DATAFRAMES IN PANDAS

```python
#creating df from dictionary of list
data={
    'name': ['sahil' ,'sagar ','ted'],
    'Age':[24,23,22],
    'City':['bengaluru','Belagav','california']

}
df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
#creating  a dataframe from a list of dictionaries

data=[
    {'name':'darshan','age':28,'city':'tumkur'},
    {'name':'tushar ','age':23,'city':'belagav'},
    {'name':'mike ','age':25,'city':'new york'}
]
df=pd.DataFrame(data)
print(df)
print(type(df))
```
