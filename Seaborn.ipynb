```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')#loads the dataset which are inbuilt
tips
```

```python
import matplotlib.pyplot as plt
```

```python
#scatter plot
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of Total Bill vs Tip")
plt.show()
```

```python
#line plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line plot of Total Bill by Size")
plt.show()
```

```python
#Categorical variable plots
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar plot of Total Bill by Day")
plt.show()
```

```python
#box plot
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box plot of Total Bill by Day")
plt.show()
```

```python
#violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("Violin plot of Total Bill by Day")
plt.show()
```

```python
#histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)#kde=true creates the smoothening/creating of curve
 #kde(kernel density estimator)
```

```python
#KDE  PLOT
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
#pairplott
sns.pairplot(tips)# creates the graph in relation with each of the elements
```

```python
#heat map
correlation=tips[['total_bill','tip','size']].corr()
correlation
```

```python
sns.heatmap(correlation,annot=True,cmap='coolwarm')
```
