```python
#merging and joining Dataframe
df1=pd.DataFrame({'key':['A','B','C'],'value':[1,2,3]})
df2=pd.DataFrame({'key':['A','B','D'],'value':[3,4,5]})
```

```python
pd.merge(df1,df2,on='key',how='inner')#inner join
```

```python
pd.merge(df1,df2,on='key',how='left')#left join
```
